Hace unos meses se inauguró un nuevo restaurante, y el equipo directivo quiere optimizar el espacio para sentarse en función del tamaño de los grupos que acuden con más frecuencia. Una noche, hay 10 grupos de personas esperando para sentarse en el restaurante, pero en lugar de que se llamen en el orden en que llegaron, se llamarán de forma aleatoria. En este ejercicio, investigarás la probabilidad de que se elija en primer lugar grupos de diferentes tamaños. El DataFrame restaurant_groups contiene los datos de cada uno de los diez grupos.

Recuerda que el valor esperado puede calcularse multiplicando cada resultado posible por su probabilidad correspondiente y tomando la suma. Los datos de restaurant_groups están disponibles. Se ha cargado pandas como pd, numpy como np y matplotlib.pyplot como plt.

### Instrucciones 2/4

Cuenta el número de cada group_size en restaurant_groups y luego divídelo entre el número de filas en restaurant_groups para calcular la probabilidad de seleccionar aleatoriamente un grupo de cada tamaño. Guárdalo como size_dist.
Restablece el índice de size_dist.
Cambia el nombre de las columnas de size_dist a group_size y prob.

```python
# Create probability distribution
size_dist = restaurant_groups['group_size'] / restaurant_groups.size

# Reset index and rename columns
size_dist = size_dist.reset_index()
size_dist.columns = ["group_size", "prob"]

print(size_dist)
```

### Instrucciones 3/4

Calcula el valor esperado de size_dist, que representa el tamaño esperado del grupo, multiplicando group_size por prob y obteniendo la suma.

```python
# Create probability distribution
size_dist = restaurant_groups['group_size'].value_counts() / restaurant_groups.shape[0]
# Reset index and rename columns
size_dist = size_dist.reset_index()
size_dist.columns = ['group_size', 'prob']

# Calculate expected value
expected_value = np.sum(size_dist['group_size']*size_dist['prob'])
print(expected_value)
```

### Instrucciones 4/4

Calcula la probabilidad de elegir aleatoriamente un grupo de 4 o más personas haciendo subconjuntos para los grupos con un tamaño de 4 o más y sumando las probabilidades de seleccionar esos grupos.

```python
# Create probability distribution
size_dist = restaurant_groups['group_size'].value_counts() / restaurant_groups.shape[0]
# Reset index and rename columns
size_dist = size_dist.reset_index()
size_dist.columns = ['group_size', 'prob']

# Expected value
expected_value = np.sum(size_dist['group_size'] * size_dist['prob'])

# Subset groups of size 4 or more
groups_4_or_more = size_dist[size_dist['group_size']>3]

# Sum the probabilities of groups_4_or_more
prob_4_or_more = groups_4_or_more['prob'].sum()
print(prob_4_or_more)
```

```python
# Min and max wait times for back-up that happens every 30 min
min_time = 0
max_time = 30
```

### Instrucciones 2/4

Importa uniform de scipy.stats, calcula la probabilidad de que Amir tenga que esperar **menos de 5 minutos** y almacénala en una variable llamada prob_less_than_5.

```python
# Min and max wait times for back-up that happens every 30 min
min_time = 0
max_time = 30

# Import uniform from scipy.stats
from scipy.stats import uniform

# Calculate probability of waiting less than 5 mins
prob_less_than_5 = uniform.cdf(5, min_time, max_time)
print(prob_less_than_5)
```

### Instrucciones 3/4

Calcula la probabilidad de que Amir tenga que esperar **más de 5 minutos** y almacénala en una variable llamada prob_greater_than_5.

```python
# Min and max wait times for back-up that happens every 30 min
min_time = 0
max_time = 30

# Import uniform from scipy.stats
from scipy.stats import uniform

# Calculate probability of waiting more than 5 mins
prob_greater_than_5 = 1- uniform.cdf(5, min_time, max_time)
print(prob_greater_than_5)
```

### Instrucciones 4/4

```python
# Min and max wait times for back-up that happens every 30 min
min_time = 0
max_time = 30

# Import uniform from scipy.stats
from scipy.stats import uniform

# Calculate probability of waiting 10-20 mins
prob_between_10_and_20 = uniform.cdf(20, min_time, max_time) - uniform.cdf(10, min_time, max_time)
print(prob_between_10_and_20)
```

### Instrucciones 1/3

¿Cuál es la probabilidad de que Amir **cierre los 3 acuerdos** en una semana? Guárdalo como prob_3.

```python
# Probability of closing 3 out of 3 deals
prob_3 = binom.pmf(3, 3, 0.3)

print(prob_3)
```

### Instrucciones 2/3

¿Cuál es la probabilidad de que Amir cierre 1 acuerdo o menos en una semana? Guárdalo como prob_less_than_or_equal_1.

```python
# Probability of closing <= 1 deal out of 3 deals
prob_less_than_or_equal_1 = binom.cdf(1, 3, 0.3)

print(prob_less_than_or_equal_1)
```

### Instrucciones 3/3

¿Cuál es la probabilidad de que Amir cierre más de 1 acuerdo? Guárdalo como prob_greater_than_1.

```python
# Probability of closing > 1 deal out of 3 deals
prob_greater_than_1 = 1- binom.cdf(1,3,0.3)

print(prob_greater_than_1)
```
